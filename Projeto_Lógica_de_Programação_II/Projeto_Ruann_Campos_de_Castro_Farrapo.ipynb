#### 2. Implemente uma função que solicite ao usuário a ação desejada (Incluir ou Excluir) um item da lista. 

    Informações sobre a venda que devem constar:

```python
# Inicializando a estrutura de dados que irá armazenar as vendas realizadas. 
# Uma lista que receberá um dicionário para cada registro de venda feita. Uma lista de dicionários de regsitros de vendas.
lista_vendas = []
# Importando as bibliotecas que possuem métodos que foram utilizados no decorrer da construção do projeto 
# Faço explanações sobre as mesmas quando as utilizei na construção do projeto
import csv
from ast import literal_eval
from datetime import datetime
```

```python
# Função auxiliar para tratamento da Data da Venda quando o usuário fizer a inclusão de venda.
 # Utilizando datetime, transformo o input recebido do usuário em tipo data para ficar um registro mais fiel de data.
# Utilizei um try except na construção da função auxiliar, para quando o usuário passar a data, caso o mesmo digite um formato inválido
# Caso digite errado, será informado e requerido que o usuário digite novamente .
def tratamento_data(data_venda):
    try:
        data_venda = datetime.strptime( data_venda, '%d/%m/%Y')
    except:
        print('Formato de Data inválido. Informe novamente a data com o padrão DD/MM/YYYY!')
        data_venda = input("Informe a data de venda(DD/MM/AAAA):\n ")
        data_venda = datetime.strptime(data_venda, '%d/%m/%Y')
    #Após todo esse processo transformo os dados de data em str com método strftime
    #Assim, fica mais fácil o acesso e manipulação desses dados de registros das datas de vendas.
    data_venda = data_venda.strftime('%d/%m/%Y')
    return data_venda

# Função auxiliar para tratamento da inclusão do CPF do comprador quando o usuário fizer a inclusão de venda.
def tratamento_cpf():
    cpf = input('Informe o CPF do comprador(somente números):\n ')
    # Um while para tratar o erro de o usuário não repassar os 11 dígitos necessários do cpf
    while len(cpf) != 11:
        cpf = input('Quantidade de Dígitos Incorreto. CPF possui 11 números. Por favor, informe novamente(somente números): ')
    cpf = '{}.{}.{}-{}'.format(cpf[:3], cpf[3:6], cpf[6:9], cpf[9:])
    # Cada cpf passado é salvo em uma tupla, protegendo esses dados pessoais de mudanças indevidas
    cpf = tuple(cpf.split())
    return cpf
```

```python
# Definindo a função Acao_Venda que irá receber do usuário a opção 'Incluir' ou 'Excluir' para a inclusão ou exclusão de uma venda
def Acao_Venda(acao:str,**informacao_opcional):
    if acao.lower() == 'incluir':
# Cada requisição do usuário, é incrementado +1 para o novo ID de venda, somando sempre +1 com a quantidade de elementos da lista
        id_v = len(lista_vendas) + 1
# Para os nomes, quantidades e valores dos produtos, que podem ser mais de 1, utilizei o método map para interar em cada input feito para o usuário
# Utilizei o split, para que se caso seja passado mais de um produto por exemplo. 
# É considerado um novo valor de entrada quando o usuário separa utilizando a vírgula. Salvo as entradas em uma lista com o list
        nome_p = list(map(str,input('Informe o(s) nome(s) do(s) produto(s) vendido(s). Sendo mais de 1, separe-os por vírgulas:\n ')
                        .split(','))) 
        quantidade_p = list(map(int,input('Informe a(s) quantidade(s) do(s) produto(s) vendido(s). Sendo mais de 1, separe-as por vírgulas:\n')
                              .split(',')))
        valor_p = list(map(float,input('Informe o(s) valor(es) do(s) produto(s) vendido(s). Sendo mais de 1, separe-os por vírgulas:\n')
                              .split(',')))
        nome_vendedor = str(input('Informe o nome do vendedor:\n '))
        cpf_comprador = (tratamento_cpf()) #chamando função auxiliar do tratamento de cpf
        data_venda = input('Informe a data de venda(DD/MM/AAAA):\n ')
        data_venda = tratamento_data(data_venda) #chamando função auxiliar do tratamento de data
        informacao_opcional = {input('Alguma informação opcional? Informe aqui. Caso não haja, aperte somente enter!\n ')}
        # Adicionando na lista de vendas os dicionários de cada um dos registros de vendas incluídos
        lista_vendas.append(
            {
            'ID_VENDA' : id_v,
            'Nome do(s) Produto(s)': nome_p,
            'Quantidade(s) do(s) Produtos': quantidade_p,
             'Valor(es) do(s) Produtos R$': valor_p,
             'Nome Vendedor': nome_vendedor,
             'CPF do Comprador' : cpf_comprador,
             'Data de Venda' : data_venda,
             'Informação Opcional' : informacao_opcional
             }
        ) 
# Para a apresentação de todos os registros da lista de venda após atualização de inclusão, eu utilizei list comprehension
        print('\nInclusão de venda bem sucedida! Segue abaixo a lista de vendas atualizada:\n\n Vendas:',
              [vendas for vendas in lista_vendas])
    elif acao.lower() == 'excluir':
        id_venda = int(input('Informe o ID da venda que você deseja excluir: '))
        for i in range(len(lista_vendas)):
            if lista_vendas[i]['ID_VENDA'] == id_venda:
                lista_vendas[i] = {'ID_VENDA':id_venda,'Nome do(s) Produto(s)':'VENDA CANCELADA'}
                print('Venda cancelada!')
# Para a apresentação de todos os registros da lista de venda após atualização de exclusão, eu utilizei list comprehension
                print('Segue abaixo a lista de vendas atualizada:\n\n Vendas:', 
                      [vendas for vendas in lista_vendas])
            if i == len(lista_vendas)-1 and lista_vendas[i]['ID_VENDA'] != id_venda:
                print('ID não encontrado! Por favor, tente novamente!')
    else:
        print('Opção Incorreta! Declare novamente a ação desejada.')
```

```python
# Definindo um dicionário que corresponde aos meses e depois fazendo um list comprehension para definir as chaves do dicionário
# Sendo as chaves correspondentes de 1:'01', para que assim, seja padronizado a forma que o usuário escolhe o mês da busca
# Utilizo a função lambda para gerar uma função que receba essa chave e devolva o mês correspondente de 1('01') a 12('12')
mes_num = ['01','02','03','04','05','06','07','08','09','10','11','12']
mes_dict_num = {k: mes_num for k, mes_num in enumerate(mes_num, start=1)}
num_mes = lambda mes_usuario: mes_dict_num[mes_usuario]
# Função lambda auxiliar usada para a apresentação dos resultados. Passo os nomes dos meses e defino suas chaves de 1 a 12 também.
# A relação é para que quando for apresentar o resultado, possa ser apresentado também o nome do mês escolhido
# É feito o mesmo procedimento da num_mes
mes_nome = ['Janeiro','Fevereiro','Março','Abril','Maio','Junho','Julho',
            'Julho','Agosto','Setembro','Outubro','Novembro','Dezembro']
mes_dict_nome = {k: mes_nome for k, mes_nome in enumerate(mes_nome, start=1)}
num_nome = lambda mes_usuario: mes_dict_nome[mes_usuario]
# Ao se chamar essa função, explica-se ao usuário a padronização da busca por mês.
# Depois pede-se ao usuário que ele informe o mês desejado de acordo com a padronização já explanada acima.
# Definindo uma busca padronizada para o usuário em que o mesmo passa de 1 a 12. 
def mes_usu():
    print('''Para padronização de busca, informe o mês desejado repassando um número entre 1 e 12. Segue abaixo as correspondências: 
    [1: Janeiro, 2: Fevereiro, 3: Março, 4: Abril, 5: Maio, 6: Junho, 7: Julho, 8: Agosto, 9: Setembro, 10: Outubro, 
    11: Novembro, 12: Dezembro]
    ''')
    mes_usuario =  int(input('Informe o mês desejado(1 a 12): '))
    while mes_usuario < 1 or mes_usuario > 12:
        mes_usuario =  int(input('Entrada Inválida! Informe o mês novamente(1 a 12): '))
    return mes_usuario
# Função auxiliar que pergunta ao usuário se ele quer passar o ano na busca e recebe sua resposta
def mensagem_ano():
    print(''' 'Você deseja escolher o ANO para ver o vendedor com maior faturamento correspondente? 
    Digite 'Sim' caso queira ver o ANO específico.
    Digite 'Não' caso queira ver período geral englobando todos os anos que possuem vendas registradas
    Digite abaixo:
    ''')
    resposta_ano = str(input())
    while resposta_ano.upper() != 'SIM' and resposta_ano.upper() != 'NÃO':
        resposta_ano = str(input('Entrada inválida! Informe novamente: '))
    return resposta_ano
# Função auxiliar que pergunta ao usuário se ele quer passar o mês na busca e recebe sua resposta
def mensagem_mes():
    print(f''' 'Você deseja escolher o MÊS para ver o vendedor com maior faturamento correspondente? 
    Digite 'Sim' caso queira ver o MÊS específico.
    Digite 'Não' caso queira ver período geral englobando todos meses que possuem vendas registradas
    Digite abaixo:
    ''')
    resposta_mes = str(input())
    while resposta_mes.upper() != 'SIM' and resposta_mes.upper() != 'NÃO':
        resposta_mes = str(input('Entrada inválida! Informe novamente: '))
    return resposta_mes
# Função auxiliar que recebe do usuário o ano desejado e também faz o tratamento caso ele passe em um formato inválido. Ex:'200'
def aux_ano():
    ano = input('Informe o ano desejado(AAAA): ')
    while len(ano) < 4 and ano == '0000':
        ano = input('Formato inválido!. Informe o ano novamente(AAAA): ')
    return ano
```

#### 3. Crie uma função que calcule o valor total e o valor médio das vendas em um determinado mês.

```python
def calc_valor_mes():
    # inicializando variáveis utilizadas na construção da função
    valor_total, count_quant, media_vendas = 0,0,0
    # Pegando mês e ano passado pelo usuário com as funções auxiliares
    mes_usuario =  mes_usu()
    ano = aux_ano()
    # Iterando para pegar cada venda registrada
    for vendas in lista_vendas:
# Aqui faço a condição para que a iteração ocorra somente nas vendas válidas. 
# Não tendo assim iteração desnecessárias nas vendas excluídas/canceladas.
        if len(vendas) != 2:
            # Aqui separo os meses e ano de cada um das vendas na iteração
            mes_venda = vendas['Data de Venda'].split('/')[-2]
            ano_venda = vendas['Data de Venda'].split('/')[-1]
# Faço a comparação se há correspondência do ano e mês da venda com o mês e ano passado pelo usuário
            if mes_venda == num_mes(mes_usuario) and ano_venda == ano:
# Aqui utilizo o sum, porque pode haver vendas com mais de um produto vendido, tendo assim mais de um valor na lista de valores
                valor_total += sum(vendas['Valor(es) do(s) Produtos R$'])
# Pego a quantidade dos produtos vendidos utilizando também o sum, porque pode haver mais de um produto
                count_quant += sum(vendas['Quantidade(s) do(s) Produtos']) 
    
    if valor_total == 0:
        print('Não há registros de vendas no período passado!')
    elif valor_total == 0 and count_quant != 0:
        print(f'O valor da venda está zerada em {num_nome(mes_usuario)}, logo o Valor Médio de Vendas também é R$0.')
    else:
# Faço o Valor Médio através do valor total dividido pela quantidade de unidades de cada um dos produtos vendidos no período de busca do usuário
        media_vendas = valor_total/count_quant
        print(f'Em relação a {num_nome(mes_usuario)} de {ano}, o Valor Total das Vendas foi de R${round(valor_total,2)}.')
        print(f'Ressalta-se que o Valor Médio das Vendas em relação a {num_nome(mes_usuario)} de {ano}, foi de R${round(media_vendas,2)}.')
```

```python
# CHAMANDO a função que calcula o valor total e o valor médio das vendas em um determinado mês.
# APRESENTANDO RESULTADO abaixo de uma requisição do usuário
calc_valor_mes()
```

#### 4. Crie uma função para descobrir quem foi o vendedor que obteve o maior faturamento do mês. A função deve permitir ao usuário definir o mês e o ano desejados.

```python
def calc_melhor_vendedor():
    # Recebendo do usuário se ele deseja passar mês em sua busca
    resposta_mes = mensagem_mes()
    resposta_ano = mensagem_ano()
    # Inicializando as listas e variáveis auxiliares que serão usadas nas condições
    aux_vend, aux_vend2 = [],[]
    valor_vendedor, valor_total = 0,0
# Fiz quatro condições! Nesta primeira é se o usuário desejar passar tanto o mês quanto o ano 
    if resposta_mes.upper() == 'SIM' and resposta_ano.upper() == 'SIM':
# Pegando mês e ano passado pelo usuário com as funções auxiliares
        mes_usuario =  mes_usu()
        ano = aux_ano()
# Iterando para pegar cada venda registrada
        for vendas in lista_vendas:
# Aqui faço a condição para que a iteração ocorra somente nas vendas válidas. 
# Não tendo assim iteração desnecessárias nas vendas excluídas/canceladas.
# Esses passos se repetem em todas condições
            if len(vendas) != 2:
        # Pegando mês e ano de cada venda
                mes_venda = vendas['Data de Venda'].split('/')[-2]
                ano_venda = vendas['Data de Venda'].split('/')[-1]
        # Aqui eu pego o nome de todos os vendedores que tem no período passado
                if mes_venda == num_mes(mes_usuario) and ano_venda == ano:
                    aux_vend.append({'Nome_Vendedor':vendas['Nome Vendedor']})
                    # Eu salvo todas as vendas nessa auxiliar para iterar sobre ela depois
                    aux_vend2.append(vendas)
        #Aqui com este método de set, eu pego isoladamente e individualmente o nome de cada vendedor
        aux_vend = list(set(val for dic in aux_vend for val in dic.values()))
# Aqui faço essa condição direta já pra caso não tenha vendedor, informar que não há vendas registradas, caso tenha, faço todo o trabalho dentro do if
        if aux_vend != []:
# Aqui faço uma iteração pegando por cada vendedor isoladamente            
            for i in range(len(aux_vend)):
        # Faço uma função lambda e a filter para isolar e filtrar todos os registros de vendas correspontes ao vendedor  
                vendedor_func = lambda dc: dc['Nome Vendedor'] == aux_vend[i]
                vendedor_vendas = list(filter(vendedor_func, aux_vend2))
    # Após filtragem, itero sobre os registros do vendedor e vou somando todo seu faturamento
                for vendas in vendedor_vendas:
                    valor_total += sum(vendas['Valor(es) do(s) Produtos R$'])
    # Como valor_vendedor inicia com 0, na primeira iteração salva o primeiro valor do vendedor e vai comparando com os que vem depois
    # Caso o faturamento do próximo vendedor seja maior, valor vendedor recebe o novo valor 
    # Também vou salvando o nome do vendedor com o maior faturamente em cada comparação feita
                if valor_total > valor_vendedor:
                    valor_vendedor = valor_total
                    nome_vendedor = aux_vend[i]
    # Após iterar sobre um vendedor, eu zero o valor total novamente, para que seja calculado o valor de faturamento do próximo vendedor
                valor_total = 0 
            print(f'\nO(a) vendedor(a) {nome_vendedor} apresentou o maior faturamento em {num_nome(mes_usuario)} de {ano}, sendo ele de: R${round(valor_vendedor,2)}')
        else:
            print(f'\nNão há vendas registradas no período correspondente à {num_nome(mes_usuario)} de {ano}!')
# RESSALTO QUE OS COMENTÁRIOS FEITOS NO BLOCO DE CONDIÇÃO ACIMA SERVE PARA TODOS OS OUTROS TRÊS, JÁ QUE É O MESMO PENSAMENTO
# A ÚNICA COISA QUE MUDA É A CONDIÇÃO DE CADA BLOCO E O RESULTADO DE ACORDO COM O DESEJO DO USUÁRIO(ESCOLHENDO MÊS OU ANO OU NÃO ESCOLHENDO)
    elif resposta_mes.upper() == 'SIM' and resposta_ano.upper() == 'NÃO':
        mes_usuario =  mes_usu()
        for vendas in lista_vendas:
            if len(vendas) != 2:
                mes_venda = vendas['Data de Venda'].split('/')[-2]
                if mes_venda == num_mes(mes_usuario):
                    aux_vend.append({'Nome_Vendedor':vendas['Nome Vendedor']})
                    aux_vend2.append(vendas)
        aux_vend = list(set(val for dic in aux_vend for val in dic.values()))
        if aux_vend != []:
            for i in range(len(aux_vend)):
                vendedor_func = lambda dc: dc['Nome Vendedor'] == aux_vend[i]
                vendedor_vendas = list(filter(vendedor_func, aux_vend2))
                for vendas in vendedor_vendas:
                    valor_total += sum(vendas['Valor(es) do(s) Produtos R$'])
                if valor_total > valor_vendedor:
                    valor_vendedor = valor_total
                    nome_vendedor = aux_vend[i]
                valor_total = 0 
            print(f'\nO(A) vendedor(a) {nome_vendedor} apresentou o maior faturamento no período geral em que há vendas registradas no mês de {num_nome(mes_usuario)}!')
            print(f'Seu faturamento foi de: R${round(valor_vendedor,2)}')
        else:
            print(f'\nNão há vendas registradas no período correspondente ao mês de {num_nome(mes_usuario)}!')
    
    elif resposta_mes.upper() == 'NÃO' and resposta_ano.upper() == 'SIM':
        ano = aux_ano()
        for vendas in lista_vendas:
            if len(vendas) != 2:
                ano_venda = vendas['Data de Venda'].split('/')[-1]
                if ano_venda == ano:
                    aux_vend.append({'Nome_Vendedor':vendas['Nome Vendedor']})
                    aux_vend2.append(vendas)
        aux_vend = list(set(val for dic in aux_vend for val in dic.values()))
        if aux_vend != []:
            for i in range(len(aux_vend)):
                vendedor_func = lambda dc: dc['Nome Vendedor'] == aux_vend[i]
                vendedor_vendas = list(filter(vendedor_func, aux_vend2))
                for vendas in vendedor_vendas:
                    valor_total += sum(vendas['Valor(es) do(s) Produtos R$'])
                if valor_total > valor_vendedor:
                    valor_vendedor = valor_total
                    nome_vendedor = aux_vend[i]
                valor_total = 0 
            print(f'\n{nome_vendedor} está de parabéns! Ele(a) apresentou o maior faturamento quando se leva em conta todas as vendas registradas em {ano}!')
            print(f'Seu faturamento foi de: R${round(valor_vendedor,2)}')
        else:
            print(f'\nNão há vendas registradas no período correspondente ao ano de {ano}!')
        
    else:
        for vendas in lista_vendas:
            if len(vendas) != 2:
                aux_vend.append({'Nome_Vendedor':vendas['Nome Vendedor']})
                aux_vend2.append(vendas)
        aux_vend = list(set(val for dic in aux_vend for val in dic.values()))
        if aux_vend != []:
            for i in range(len(aux_vend)):
                vendedor_func = lambda dc: dc['Nome Vendedor'] == aux_vend[i]
                vendedor_vendas = list(filter(vendedor_func, aux_vend2))
                for vendas in vendedor_vendas:
                    valor_total += sum(vendas['Valor(es) do(s) Produtos R$'])
                if valor_total > valor_vendedor:
                    valor_vendedor = valor_total
                    nome_vendedor = aux_vend[i]
                valor_total = 0 
            print(f'\n{nome_vendedor} é o(a) RECORDISTA! Ele(a) apresentou o maior faturamento desde que as vendas começaram a ser registradas!')
            print(f'Seu faturamento recorde foi de: R${round(valor_vendedor,2)}')
        else:
            print(f'\nNão há nenhuma venda registrada!')
```

```python
def calc_produto_vendido():
    # Pegando mês e ano passado pelo usuário com as funções auxiliares
    mes_usuario =  mes_usu()
    ano = aux_ano()
# Inicializando a lista que receberá a quantidade de produto mais vendido do mês e também a variável que receberá o nome do produto mais vendido
    count_quant = [0]
    nome_produto = ''
    
    for vendas in lista_vendas:
        if len(vendas) != 2:
# Aqui faço a condição para que a iteração ocorra somente nas vendas válidas. 
# Não tendo assim iteração desnecessárias nas vendas excluídas/canceladas.
            mes_venda = vendas['Data de Venda'].split('/')[-2]
            ano_venda = vendas['Data de Venda'].split('/')[-1]
# Faço a comparação se há correspondência do ano e mês da venda com o mês e ano passado pelo usuário
            if mes_venda == num_mes(mes_usuario) and ano_venda == ano:
# Faço essa condição para iterar sobre as vendas que possuem somente um produto, como também das que posssuem mais de um produto
# Assim eu itero sobre cada registro de venda e vou salvando os produtos que possuem maior quantidade de unidades vendidas
# Vou fazendo a comparação em cada iteração e vou salvando a maior quantidade no count_quant e o nome do produto em nome_produto
                if len(vendas['Nome do(s) Produto(s)']) == 1:
                    if count_quant <  vendas['Quantidade(s) do(s) Produtos']:
                        count_quant = vendas['Quantidade(s) do(s) Produtos']
                        nome_produto = vendas['Nome do(s) Produto(s)']
# Caso a venda apresente mais de um produto, itero sobre cada um dos valores da venda
                else:
                    for i in range(len(vendas['Nome do(s) Produto(s)'])):
                        if count_quant <  [vendas['Quantidade(s) do(s) Produtos'][i]]:
                            count_quant = [vendas['Quantidade(s) do(s) Produtos'][i]]
                            nome_produto = vendas['Nome do(s) Produto(s)'][i] 

    if count_quant == [0]:
        print(f'Não houve vendas em relação ao mês de {num_nome(mes_usuario)}!')
    else:
        print(f'O produto {nome_produto[0]} foi mais vendido no mês de {num_nome(mes_usuario)}. Com a venda de {count_quant[0]} unidades!')
```

#### 7. Crie uma função para converter os valores dos produtos em Dólar e salvar em um novo arquivo (Utilizando Função geradora). 

```python
# Fazendo a leitura do arquivo com os dados da lista de venda com o with open para utilizá-lo na função de conversão dos valores em dólares
with open('Dados_Lista_de_Vendas.csv', 'r') as arquivo:
    planilha = csv.reader(arquivo, delimiter=',', lineterminator='\n')
    # Repassando os dados de vendas salvos no arquivo para a variável tabela que será passada na função de conversão.
    tabela = list(planilha)
```

```python
def conversor_dolar(tabela):
    # Inicializo a lista que receberá os dados atualizados de vendas
    lista_vendas_dolar = []
# Faço esse passo de iteração para relacionar as chaves com os valores correspondentes
# Porque os dados advindos do arquivo csv que estão armazenados na tabela, vieram por exemplo com as chaves(colunas), como fossem um dicionário de venda
# Para a reorganização dos dados, e consequentemente facilitar assim, as ações necessárias, faço essa nova lista
# Reorganizando assim cada chave com os seus devidos valores. Mudei o cifrão da chave de valores, para corresponder com o dólar.
    for i in range(1,len(tabela)):
        lista_vendas_dolar.append(
        {
            tabela[0][0]:tabela[i][0],
            tabela[0][1]:tabela[i][1],
            tabela[0][2]:tabela[i][2],
            'Valor(es) do(s) Produtos em U$':tabela[i][3],
            tabela[0][4]:tabela[i][4],
            tabela[0][5]:tabela[i][5],
            tabela[0][6]:tabela[i][6],
            tabela[0][7]:tabela[i][7]
        }
        )
# Inicilizando lista auxiliar
    aux_lista = []
    for vendas in lista_vendas_dolar:
# O método literal_eval da biblioteca ast foi utilizado para retirar as aspas de string que vieram nos valores da vendas.
# Dando um exemplo, os Valores retornaram após a leitura do arquivo csv da seguinte maneira: '[12]'
# Foi utilizado após leitura dos dados de vendas do arquivo 'Dados_Lista_de_Vendas.csv' e utilização dos mesmos para a conversão em dólar
        if vendas['Nome do(s) Produto(s)'] != 'VENDA CANCELADA':
            vendas['Valor(es) do(s) Produtos em U$'] = list(literal_eval(vendas['Valor(es) do(s) Produtos em U$']))
        aux_lista.append(vendas)
# Aqui passo para a lista de vendas, os valores já sem essas aspas de string advindas dos dados de vendas.
# Possibilitando a realização de operações matemáticas que serão necessárias com os valores 
    for i in range(len(lista_vendas_dolar)):
        lista_vendas_dolar[i]['Valor(es) do(s) Produtos em U$'] = aux_lista[i]['Valor(es) do(s) Produtos em U$']
    
    cotacao_dolar = 4.73 # Cotação apurada em: 31/07/2023 
# Inicializando lista auxiliar novamente. Esse passo é necessário porque a mesma precisa estar zerada novamente para ser usada abaixo.
    aux_lista = []
# Inicializando variáveis
    valor, valor_convertido = 0,0
    for i in range(len(lista_vendas_dolar)):
# Aqui faço a condição para que a iteração ocorra somente nas vendas válidas. 
# Não tendo assim iteração desnecessárias das vendas excluídas/canceladas.
        if lista_vendas_dolar[i]['Nome do(s) Produto(s)'] != 'VENDA CANCELADA':
# Faço essa condição para fazer a conversão tanto das vendas que possui somente um valor de um produto, como também das que posssuem mais de uma valor de produto
            if len(lista_vendas_dolar[i]['Valor(es) do(s) Produtos em U$']) == 1:
                valor = lista_vendas_dolar[i]['Valor(es) do(s) Produtos em U$']
# Aqui pego o valor do produto e faço a multiplicação com a taxa de câmbio do dólar
                valor_convertido = valor[0]*cotacao_dolar
# Aqui adiciono o novo valor convertido na lista de vendas em dólar. Valores esses que serão salvos no arquivo já na função
                lista_vendas_dolar[i]['Valor(es) do(s) Produtos em U$'] = [round(valor_convertido,2)]
            else:
# Caso a venda apresente mais de um valor, itero sobre cada um dos valores da venda
                for y in lista_vendas_dolar[i]['Valor(es) do(s) Produtos em U$']:
# Faço a conversão para cada um dos valores
                        valor_convertido = y*cotacao_dolar
# Adiciono em uma lista auxiliar, que depois eu a passo para a lista de vendas principal. Passo para a venda correspondente tendo base a iteração
# Passando assim para a lista de vendas em dólar, a lista de valores para cada um dos produtos do registro da venda 
                        aux_lista.append(round(valor_convertido,2))
                lista_vendas_dolar[i]['Valor(es) do(s) Produtos em U$'] = aux_lista

#Definindo as chaves como colunas
    colunas = lista_vendas_dolar[0].keys()    
# Definindo o nome do arquivo e salvando o mesmo com os valores convertidos em dólares.
    with open('Lista_de_Vendas_Conversão_Dólares.csv', 'w') as arquivo:
        planilha = csv.DictWriter(arquivo, fieldnames=colunas, lineterminator='\n')
        planilha.writeheader()
        planilha.writerows(lista_vendas_dolar)
```
